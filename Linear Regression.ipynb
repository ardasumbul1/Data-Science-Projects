```python
#Created 5 different feature set to use regression algorithm and calculate their performance

set1 = df[["Alcohol","BMI","Diphtheria","GDP","Thinness 1-19 years","Schooling","Developing","HIV/AIDS"]]
set2 = df[["Adult Mortality","Percentage expenditure","Polio","GDP","Thinness 5-9 years","Income composition of resources","Developed","Schooling"]]
set3 = df[["Alcohol","Adult Mortality","Percentage expenditure","BMI","GDP","Polio","Diphtheria","Developing"]]
set4 = df[["Thinness 1-19 years","Thinness 5-9 years","Income composition of resources","Schooling","Developed","GDP","HIV/AIDS","Polio"]]
set5 = df[["Adult Mortality","BMI","Diphtheria","Income composition of resources","Thinness 1-19 years","Schooling","Developing","HIV/AIDS"]]
yset = df.iloc[:,0]
```

```python
#Split these sets as train and test sets

from sklearn.model_selection import train_test_split
```

```python
x_train_set1, x_test_set1,y_train_set1,y_test_set1=train_test_split(set1,yset,test_size=0.2, random_state=147)
x_train_set2, x_test_set2,y_train_set2,y_test_set2=train_test_split(set2,yset,test_size=0.2, random_state=147)
x_train_set3, x_test_set3,y_train_set3,y_test_set3=train_test_split(set3,yset,test_size=0.2, random_state=147)
x_train_set4, x_test_set4,y_train_set4,y_test_set4=train_test_split(set4,yset,test_size=0.2, random_state=147)
x_train_set5, x_test_set5,y_train_set5,y_test_set5=train_test_split(set5,yset,test_size=0.2, random_state=147)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
#Created 5 different scaler to scale sets

scaler1=MinMaxScaler(feature_range=(0,1))
scaler2=MinMaxScaler(feature_range=(0,1))
scaler3=MinMaxScaler(feature_range=(0,1))
scaler4=MinMaxScaler(feature_range=(0,1))
scaler5=MinMaxScaler(feature_range=(0,1))
```

```python
#Scaled trainin and test sets of xs's

scaled_x1 = scaler1.fit_transform(x_train_set1)
scaled_x2 = scaler2.fit_transform(x_train_set2)
scaled_x3 = scaler3.fit_transform(x_train_set3)
scaled_x4 = scaler4.fit_transform(x_train_set4)
scaled_x5 = scaler5.fit_transform(x_train_set5)

scaled_x1_test = scaler1.transform(x_test_set1)
scaled_x2_test = scaler2.transform(x_test_set2)
scaled_x3_test = scaler3.transform(x_test_set3)
scaled_x4_test = scaler4.transform(x_test_set4)
scaled_x5_test = scaler5.transform(x_test_set5)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#Used linear regression model to whole sets and estimated corresponding life expectancies
#Esitmated life expectancies compared with its' real values and different kinds of errors are calculated 
#according to these estimations

print("SET 1")
model = LinearRegression()
model.fit(scaled_x1,y_train_set1)

print("coefficient:")
print(model.coef_)
print("\ninterception point:")
print(model.intercept_)

y1_predict=model.predict(scaled_x1_test)
score=model.score(scaled_x1_test,y_test_set1)
print("\nscore:")
print(score)
```

```python
print("SET 1")
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y1_predict,y_test_set1)
print("\nMean squared error:")
print(mse)

from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y1_predict,y_test_set1)
print("\nMean absolute error:")
print(mae)

rmse = math.sqrt(mean_squared_error(y1_predict,y_test_set1))
print("\nRoot mean squared error:")
print(rmse)

from sklearn.metrics import r2_score
r2 = r2_score(y1_predict,y_test_set1)
print("\nR2:")
print(r2)
```
