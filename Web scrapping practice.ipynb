```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
items = []
for i in range(1,5):
    html_text = requests.get('https://www.allrecipes.com/search/results/?search=mexican/&page='+str(i)).text
    soup1 = BeautifulSoup(html_text, 'lxml')
    content_box1 = soup1.find_all('h3', class_= "card__title elementFont__resetHeading")
    for j in content_box1:
        items.append(j.text.replace("\n","").replace("  ","").replace(" ","-").replace("&","").replace("--","-").replace("'","").replace("(","").replace(")","").lower())
```

```python
output = pd.DataFrame()

for recipe_name in items:
    
    html_text = requests.get('https://www.allrecipes.com/recipe/'+ recipe_name).text
    soup = BeautifulSoup(html_text, 'lxml')
    content_box = soup.find('div', class_= "ng-dialog-content")
    time_info = soup.find('div', class_= "recipe-meta-item-body elementFont__subtitle").text
    calories = content_box.find('div', class_= "nutrition-top light-underline elementFont__subtitle").text
    recipe_name = content_box.find('h3', class_= "recipe-title elementFont__leader").text
    nut_list = {}
    nut_list["name"] = recipe_name
    nutrition_list = content_box.find_all('div', class_= "nutrition-row")

    
    ingredients_section = soup.find("ul", class_= "ingredients-section")
    ingredients = ingredients_section.find_all('li', class_= "ingredients-item")
    ingredients_list = []
    
    rec_descr = soup.find('p', class_= "margin-0-auto").text
    nut_list["description of recipe"] = rec_descr


    for ingredient in ingredients:
        ingredient_name = ingredient.find("span", class_= "ingredients-item-name elementFont__body").text
        ingredients_list.append(ingredient_name)
    str1 = " "
    ingredients_list = str1.join(ingredients_list)
    nut_list["ingredients"] = ingredients_list

    for nutrition in nutrition_list:
        nutri = nutrition.find('span', class_= "nutrient-name")
        nutri_amount = nutri.find('span',class_= "nutrient-value").text
        nut_list[nutri.span.text] = nutri_amount


    output = output.append(nut_list, ignore_index=True)
```
